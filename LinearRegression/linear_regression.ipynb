```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split   # para dividir los datos
from sklearn.linear_model import LinearRegression    # es el modelo a usar
from sklearn.metrics import mean_squared_error, r2_score   # para ver los resultados
```

**CON DATOS DE DESEMPEÑO DE ESTUDIANTES SECUNDARIO - UNIVERSIDAD**

```python
data = pd.read_csv('data.csv')

x = data[['SAT']]
y = data['GPA']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = model.predict(X_test)

print(y_pred)
```

```python
plt.scatter(X_test, y_test, color='blue', label='Actual Data')
plt.plot(X_test, y_pred, color='red', label='Regression Line')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
plt.show()
```

**CON DATOS DE METACRITIC GAMES**

```python
data = pd.read_csv('games.csv')

data = data[data['user_score'] != 'tbd']
data['user_score'] = data['user_score'].astype(float)  # Convertir user_score a float
data = data.sort_values('user_score')

x = data[['user_score']]
y = data['metascore']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
plt.scatter(x_test, y_test, color='blue', label='Actual Data')
plt.plot(x_test, y_pred, color='red', label='Regression Line')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
plt.show()
```
